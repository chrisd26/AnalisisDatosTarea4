```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('C:/Users/chris/Downloads/1 Cursos  UNAD/52 ANÁLISIS DE DATOS/Tarea 4/Python_DS_Covid/Cleaned-Data.csv')
df.sample(10)
```

```python
explicativas= df.drop(columns='Country')
objetivo=df.Country
```

```python
model= DecisionTreeClassifier(max_depth=3)
```

```python
model.fit(
X=explicativas, y=objetivo)
```

```python
plt.figure(figsize=(14, 8))
plot_tree(decision_tree=model, feature_names=explicativas.columns, filled=True, fontsize=8);
```
